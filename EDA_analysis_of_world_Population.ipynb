```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Function to import data
def wrangle (filename):
    df = pd.read_csv(filename)
    
    return df
```

```python
# To sort the data by area size
df.sort_values(by="Area (km²)", ascending = False).head()
```

```python
# Sepparating the numerical and categoricl values
df_numerical = df.select_dtypes(include=["float64", "int64"])
df_categorical = df.select_dtypes(include=["object"])
```

```python
# Compute the correlation matrix
correlation = df_numerical.corr()
correlation
```

```python
# To create a heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation, annot= True)
plt.show()
```

```python
df.columns
```

```python
# Group by "Continent" and calculate the mean of numeric columns
mean_by_continent = df_numerical.groupby(df["Continent"])[
    ['1970 Population','1980 Population', '1990 Population', '2000 Population', 
     '2010 Population', '2015 Population', '2020 Population', 
     '2022 Population']
].mean().sort_values(by="2022 Population", ascending=False)

# Display the result
mean_by_continent
```

```python
# To stripe down  a specific continent
df[df["Continent"].str.contains("Oceania")]
```

```python
# The distribution of the numerical data
df_numerical.hist(figsize=(16,20),bins=10, xlabelsize=8, ylabelsize=8);
```

```python
df_transpose=mean_by_continent.transpose()
df_transpose
```

```python
#Transposing the dataset
df_transpose.plot()

# Labelling the axi
plt.xlabel("Year", fontsize=12)
plt.ylabel("Frequecy", fontsize=12)

# Adding a title
plt.title("Population Growth Statistics worldwide", fontsize=14)

# Customize the x-axis tick labels to avoid overlap 
plt.xticks(rotation=45, ha='right');
plt.show()
```
